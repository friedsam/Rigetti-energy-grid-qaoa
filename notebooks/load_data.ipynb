```python
from pathlib import Path
import pandas as pd
import networkx as nx

# Find project root from either the repo root or the notebooks folder.
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
DATA_PATH = PROJECT_ROOT / "data"

print("Notebook cwd:", Path.cwd())
print("Project root:", PROJECT_ROOT)
print("Data folder exists:", DATA_PATH.exists())

parquet_files = sorted(DATA_PATH.glob("*.parquet"), key=lambda path: path.stat().st_size, reverse=True)
print("Parquet files:", parquet_files)

if not parquet_files:
    raise FileNotFoundError(f"No parquet files found in {DATA_PATH}")

parquet_file = parquet_files[0]
print("Using:", parquet_file.name)

df = pd.read_parquet(parquet_file)

G = nx.Graph()
for u, v, w in df[["node_1", "node_2", "weight"]].itertuples(index=False):
    G.add_edge(int(u), int(v), weight=float(w))

print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
print("Done.")

weights = [d["weight"] for _, _, d in G.edges(data=True)]
print("Weight min/mean/max:", min(weights), sum(weights) / len(weights), max(weights))
```
